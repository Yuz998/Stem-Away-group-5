```python
import pandas as pd
df=pd.read_csv("resources/matrix-ebc-paper-sparse.tsv",sep=('\t'))
df.head()
```

```python
df.rename(columns = {'(rc-160,igf-i)':"drug-gene", '0':'row indicies','[nsubj, reduced, dobj]':'dependency path','1861':"column indices",'1.0':"certainty"},inplace = True)
df.iloc[0] = ['(rc-160,igf-i)',0,'[nsubj, reduced, dobj]',1861 ,1.0]
df.head()
```

```python
df.dtypes
```

```python
newdf=df[['drug-gene']].copy()
newdf.head()
```

```python
newdf.drop_duplicates(subset =['drug-gene'],
                     keep = "first", inplace = True)

newdf.head()
```

```python
newdf['drug-gene'] = newdf['drug-gene'].str.replace(',', '/')
newdf.head()
```

```python
newdf.reset_index(inplace = True)

newdf.head()
```

```python
newdf=newdf.drop(['index'],axis=1)
newdf.head()
```

```python
newdf1=df[['dependency path','column indices']].copy()
newdf1=newdf1.sort_values(by = 'column indices')
newdf1.head()
```

```python
newdf1.drop_duplicates(subset =['dependency path'],
                     keep = "first",inplace = True)


newdf1=newdf1.drop(['column indices'],axis=1)
newdf1.head()
```

```python
newdf1.reset_index(inplace = True)
newdf1.head()
```

```python
newdf1=newdf1.drop(['index'],axis=1)
newdf1.head()
```

```python
data = [newdf, newdf1]

df1 = pd.concat(data, axis=1)
df1.head()
```

```python
df1.shape
```

```python
cxy_df=pd.read_csv("cXY.csv")
new_data = [df1, cxy_df]

df2 = pd.concat(new_data, axis=1)
df2
```

```python
df2=df2.drop(['Unnamed: 0'],axis=1)
df2
```

```python
# saving the dataframe
df2.to_csv('Pair_Path_Mapping.csv')
```
